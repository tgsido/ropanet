## Annotation Exploration

```python
ANNOTATION_DIR_PATH = '/Users/osida/Desktop/spr19-classes/cs231n/final_project/DeepFashion/Category and Attribute Prediction Benchmark/Anno'
anno_root = pathlib.Path(ANNOTATION_DIR_PATH)
print(anno_root)
```

```python
for item in anno_root.iterdir():
  print(item)
  print()
```

```python
for item in anno_root.iterdir():
  print(item)
  with open(item,'r') as f:
      count = 0
      for line in f:
        print(line, end='')
        if count == 5:
            break
        count +=1
  f.close()
  print()
  print()
                
```

```python
bbox_image_paths = []
bbox_filename = 'DeepFashion/Category and Attribute Prediction Benchmark/Anno/list_bbox.txt'
with open(bbox_filename,'r') as f:
      count = 0
      for line in f:
        count +=1
        if count >= 3:
            img_path = line[:line.find('.jpg') + 4].strip()
            img_path = img_path.lower()
            bbox_image_paths.append(img_path)    
f.close()

len(bbox_image_paths)
```

```python
bbox_set = set(bbox_image_paths)
all_paths_set = set(all_image_paths_trunc)

count = 0
for elem in bbox_set:
    print(elem)
    count += 1
    if count == 3:
        break
        
count = 0
for elem in all_paths_set:
    print(elem)
    count += 1
    if count == 3:
        break        
        
diff = bbox_set - all_paths_set
print("bbox_set - all_paths_set:", len(diff))
```

```python
for elem in sorted(list(diff)):
    print(elem)
    print()
```
