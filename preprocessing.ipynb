```python
import random
import shutil
all_image_paths = list(data_root.glob('*/*'))
all_image_paths = [str(path) for path in all_image_paths]

counter = 0
for src_path in all_image_paths:
    last_slash_index = src_path.rfind("/")
    #print("last_slash_index:",last_slash_index)
    src_filename = src_path[last_slash_index + 1:]
    second_to_last_slash_index = src_path.rfind("/", 0,last_slash_index-1)
    #print("second_to_last_slash_index:",second_to_last_slash_index)
    src_folder = src_path[second_to_last_slash_index + 1: last_slash_index]
    dest_filename = src_filename.lower()
    dest_folder = src_folder.lower()
    
    new_save_location = os.path.join(IMG_PREPROCESSED_DIR_PATH,dest_folder,dest_filename)
    counter += 1
    
    if os.path.exists(new_save_location):
        continue
        
    shutil.copy(src_path,new_save_location)
    if counter % 100 == 0:
        print("counter:",count)
```

```python
all_image_paths = list(data_root.glob('*/*'))
all_image_paths = [str(path) for path in all_image_paths]

all_image_preprocessed_paths = list(data_preprocessed_root.glob('*/*'))
all_image_preprocessed_paths = [str(path) for path in all_image_preprocessed_paths]

assert len(all_image_preprocessed_paths) == len(all_image_paths)
print("len(all_image_preprocessed_paths)",len(all_image_preprocessed_paths))
```

```python
def file_srt(s):
    last_slash_index = s.rfind("/")
    src_filename = s[last_slash_index + 1:]
    second_to_last_slash_index = s.rfind("/", 0,last_slash_index-1)
    src_folder = s[second_to_last_slash_index + 1: last_slash_index]
    dest_filename = src_filename.lower()
    dest_folder = src_folder.lower()
    return str(os.path.join(dest_folder,dest_filename))


all_image_preprocessed_paths = sorted(all_image_preprocessed_paths, key = lambda s: file_srt(s))
all_image_paths = sorted(all_image_paths, key = lambda s: file_srt(s))

all_paths_mod = set([file_srt(path) for path in all_image_paths])
all_paths_pre_mod = set([file_srt(path) for path in all_image_preprocessed_paths])

diff = all_paths_mod - all_paths_pre_mod
print("len(diff):",len(diff))

diff2 = all_paths_pre_mod - all_paths_mod
print("len(diff2):",len(diff2))

all_indices = list(range(len(all_image_preprocessed_paths)))
rand_indices = []
for n in range(5):
    rand_indices.append(random.choice(all_indices))

print("rand_indices:",rand_indices)
```

## Annotation Exploration

```python
ANNOTATION_DIR_PATH = '/Users/osida/Desktop/spr19-classes/cs231n/final_project/DeepFashion/Category and Attribute Prediction Benchmark/Anno'
anno_root = pathlib.Path(ANNOTATION_DIR_PATH)
print(anno_root)

print()
ANNOTATION_PREPROCESSED_DIR_PATH = '/Users/osida/Desktop/spr19-classes/cs231n/final_project/data/anno_pre'
anno_preprocessed_root = pathlib.Path(ANNOTATION_PREPROCESSED_DIR_PATH)
print(anno_preprocessed_root)

dirName = ANNOTATION_PREPROCESSED_DIR_PATH
try:
    # Create target Directory
    os.mkdir(dirName)
    print("Directory " , dirName ,  " Created ") 
except FileExistsError:
    print("Directory " , dirName ,  " already exists")
```

```python
for item in anno_root.iterdir():
  print(item)
  print()
```

```python
for item in anno_root.iterdir():
  item = str(item)
  print(item)
  anno_filename = item[item.rfind("/")+1:]
  
  new_anno_location = os.path.join(ANNOTATION_PREPROCESSED_DIR_PATH,anno_filename)
  print("new_anno_location:",new_anno_location)
  with open(new_anno_location,'w') as out_f:
    if 'cloth' not in anno_filename:
      with open(item,'r') as f:
          count = 0
          for line in f:
            count +=1
            if count < 3:
                out_f.write(line)
                continue
            #print("og line:",line, end='')
            image_portion = line[:line.rfind('.jpg') + 4]
            #print("image_portion:",image_portion)
            folder_file = file_srt(image_portion)
            #print("folder_file:",folder_file)
            full_image_path = os.path.join(IMG_PREPROCESSED_DIR_PATH,folder_file)
            if full_image_path not in all_image_preprocessed_paths:
                print(full_image_path, "not in original file")
                continue
            new_image_portion = os.path.relpath(full_image_path,DATA_DIR_PATH)
            #print("new_image_portion:",new_image_portion)
            out_f.write(line.replace(image_portion,new_image_portion))
            if count % 25000 == 0:
                print("count:",count)
          f.close()
    else:
      shutil.copy(item,new_anno_location)
    out_f.close()
    
  print()
  print()
                
```
